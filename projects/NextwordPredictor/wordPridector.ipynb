```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

**Tokenizer will tokenize the data with respective to the frequency of Unique words**

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([faqs])
len(tokenizer.word_index)
```

# **converting the sentence to numericals array**

```python
input_seq = []
for sentence in faqs.split('\n\n'):
  sentence =sentence+"<END>"
  tokenized_seq = (tokenizer.texts_to_sequences([sentence])[0])
  for i in range(1,len(tokenized_seq)):
    n_gram_seq = tokenized_seq[:i+1]
    input_seq.append(n_gram_seq)
```

so finding the maximum sentence length

```python
max_len = max([len(x) for x in input_seq])
print(max_len)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

# Pad sequences to equal length using pre-padding (adds zeros at the start)bold text

```python
padded_seq = pad_sequences(input_seq,maxlen=max_len,padding='pre')
```

```python
padded_seq
```

```python
X = padded_seq[:,:-1]
```

```python
Y = padded_seq[:,-1]
```

```python
X.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(Y,num_classes=928)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense,Input
```

```python
model = Sequential()
model.add(Input(shape=(max_len-1,)))
model.add(Embedding(928,100))
model.add(LSTM(150))
model.add(Dense(928,activation='softmax'))
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)
```
